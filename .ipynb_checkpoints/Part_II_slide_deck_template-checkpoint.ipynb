## Investigation Overview

Dans cette présentation, je vais montrer à quel points les sources de revenues peuvent être capitale pour determiner la solvabilité d'un emprunteur. Nous allons proceder en trois grandes étapes : 
* Visualiser les sources de revenues et les proportions des prêts soldés et non soldées;  
* Visualiser la rélation entre les sources de revenues et les statuts des prêts ainsi que d'autre variable;  
* visualiser la rélation entre le sources de revenues, la durée du prêt et les status des prêts.


## Dataset Overview

* Prosper est une société basée à San Francisco, en Californie, dans le secteur des prêts à des particuliers en ligne.
* Cet ensemble de données est extrais des données de Prosper. 
* Il est constitué d’environ 113,937 prêts et 81 variables, y compris le montant du prêt, le taux de l'emprunteur (ou taux d'intérêt), le statut actuel du prêt, le revenu de l'emprunteur, et bien d'autres. 
* Les données sont téléchargeable en utilisant le [lien](https://s3.amazonaws.com/udacity-hosted-downloads/ud651/prosperLoanData.csv). 
* La [documentation](https://docs.google.com/spreadsheets/d/1gDyi_L4UvIrLTEC6Wri5nbaMmkGmLQBk-Yx3z0XDEtI/edit#gid=0) du dataset permet de mieux comprendre chacun de ses variables.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
base_color = sns.color_palette()[0]
```

```python
# load in the dataset into a pandas dataframe
prosperLoanData = pd.read_csv("prosperLoanData.csv")
```

L'analyse des plages des sources de revenues nous montre que :
> * 28.25% des emprunteurs ont une source de revenues compris dans la plage 25,000-49,999 dollars;  
> * 27.25% ont une source de revenus dans la plage 50,000-74,999 dollars;  
> * 14.85% et 15.22% ont respectivement leurs sources de revenus dans 75,000-99,999 dollars et 100,000+ dollars.


```python
# Visualisations des proportions des source de revenues
plt.figure(figsize=(14, 7))
sns.countplot(data = prosperLoanData, y = "IncomeRange", color = base_color)
plt.title("Visualisation des sources de revenues");
```

```python
"""
Selection des lignes dont les status sont 'completed', 'ChagedOff' et 'Defaulted'
Nous ignorons l'issue des autres modalitées.
"""
df = prosperLoanData.query("LoanStatus=='Completed'or LoanStatus=='Chargedoff'or LoanStatus=='Defaulted'" )
```

```python
"""
Création d'une nouvelle variable nommé 'LoanDefaulted'.
Cette variable prend deux modalités : Yes or No.
La modalité Yes designe que la personne à un status soit à Defaulted soit à ChargedOff.
La modalité No désigne que la personne à un status completed
"""
df["LoanDefaulted"] = df.LoanStatus.replace(to_replace=["Defaulted", "Chargedoff"], value="Yes")
df["LoanDefaulted"] = df.LoanDefaulted.replace(to_replace=[ "Completed"], value="No")
```

```python
# Value_counts de la variable LoanDefauled
LoanDefaulted_counts = df.LoanDefaulted.value_counts()
LoanDefaulted_counts
```

* 69.10% des personnes ont fini de payés leur prêt
* 30.90% des personnes n'ont pas pu rembourser leurs prêts.

```python
plt.figure(figsize=(14, 6), facecolor="white")
plt.title("Personnes ayant manqué/reussie le payement du prêt");
plt.pie(LoanDefaulted_counts, labels = LoanDefaulted_counts.index, startangle=90 , counterclock=False, autopct="%.2f%%");
plt.legend();
```

```python
# Ordonnées le différentes sources de revenues
incomeRange = [ "Not employed", "$0", "$1-24,999", "$25,000-49,999", "$50,000-74,999", "$75,000-99,999", "$100,000+", "Not displayed"]
incomeRange_type = pd.api.types.CategoricalDtype(categories= incomeRange, ordered= True)
df.IncomeRange = df.IncomeRange.astype(incomeRange_type)
```

* Les personnes dont leurs sources de revenues dépasses $50000 sont ont en grandes majorité des maisons en leur nom. 
* Par contre les personnes dont leurs sources de revenus sont inférieures à 50 000 dollars en majorité n'ont pas de maison propre à eux. 
* Ces personnes vivent surement en location.

```python
# Visualisation des sources de revenues des personnes qui ont ou non une maison en leur non.
plt.figure(figsize=(10, 6))
sns.countplot(data = df, x = "IncomeRange", hue="LoanDefaulted");
plt.title("IncomeRange en fonction de LoanDefaulted");
plt.xticks(rotation=15);
```

Les emprunteurs ayant une source de revenue au moins égale à 100000 dollars ont beaucoup plus de chance d'avoir une importante somme d'argent de prêts. La somme maximale qu'ils peuvent avoir est de 35 000 dollars ce qui est nettement supérieur à ceux qui ont une source de revenues inférieur 10000 qui ont une somme maximale de 25 000. En conclusion Plus les sources de revenues sont élevées plus ont à beaucoup de chance d'avoir un gros montant de prêts.


```python
# Visualisation des sources de revenues et des montant reçuent
plt.figure(figsize=(10, 5))
sns.boxplot(data = df, x = "IncomeRange", y="LoanOriginalAmount", color=base_color)
plt.xticks(rotation=45)
plt.title("Sources de revenues et Somme recues");
```

Les prêts qui durent 12 mois ont une grande proportion de solvabilité quel que soit la source de revenus. Les personnes qui font les prêts sur 60 mois et qui ont une source de revenue au moins égale à 50 000 dollars ont plus de chance de soldé leurs prêts. Par contre ceux ayant une source de revenues inférieur à ce montant pour la même durée présente une forte proportion de prêts impayé. Les personnes qui prennent des prêts de 36 mois dont la source de revenu est inférieure à 75 000 dollars présente une forte chance de ne payer pouvoir rembourser leurs prêts. Par contre ceux ayant des revenus de 100 000 dollars et plus présente une faible proportion de dettes impayés.

```python
g = sns.FacetGrid(data = df, col = "Term", hue="LoanDefaulted", height=7)
g.map(sns.countplot, "IncomeRange")
g.add_legend()
g.set_xticklabels(rotation=45)
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle("Durée, source de revenue en fonction de loanDefaulted");
```
